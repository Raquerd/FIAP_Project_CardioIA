```python
import pandas as pd
import unicodedata

# 1. Função para normalizar o texto (remover acentos, aspas e padronizar minúsculas)
def normalizar_texto(texto):
    if isinstance(texto, str):
        # Remove acentos e converte para ASCII
        texto_sem_acento = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('ASCII')
        return texto_sem_acento.lower().strip('"')
    return texto

# Carregar os dados

# Definição das regras de classificaçao 
regras_risco = {
    'Alto Risco': ['falta de ar', 'tosse seca', 'cansaco excessivo'], 
    'Medio Risco': ['perda de apetite'], 
    'Baixo Risco': ['formigamento', 'zumbido', 'dor', 'sensibilidade']
}

# Função que percorre o dicionário e classifica a frase
def aplicar_classificacao(frase):
    texto = normalizar_texto(frase)
  
    for nivel, palavras in regras_risco.items():
        if any(palavra in texto for palavra in palavras):
            return nivel
            
    return "Nao Identificado"

#  Criar a coluna de risco e aplicar a função
df_phrase['risco'] = df_phrase['frases'].apply(aplicar_classificacao)

#  Exibir o resultado final
display(df_phrase)

# Salvar o resultado para a entrega da Fiap
# df_phrase.to_csv("Frases_classificadas.csv", index=False)
```
